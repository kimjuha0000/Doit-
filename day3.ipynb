```python
midwest = pd.read_csv('midwest.csv')
midwest
```

```python
midwest.head()
```

```python
midwest2 = midwest.copy()
midwest2
midwest2 = midwest.rename(columns = {'poptotal':'total','popasian':'asian'})
midwest2.head()
```

```python
midwest2.head()
```

```python
midwest2['asianpercent'] = (midwest2['asian']/midwest2['total'])*100
midwest2
midwest2['asianpercent'].plot.hist()
```

```python
asian_avg = sum(midwest2['asianpercent'])/len(midwest2['asianpercent'])
asian_avg
```

```python
midwest2['amount']=np.where(midwest2['asianpercent']>=asian_avg, 'large', 'small')
midwest2['amount'].value_counts().plot.bar()
```
