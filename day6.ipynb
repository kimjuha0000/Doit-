```python
mpg = pd.read_csv('mpg.csv')
df_mpg = mpg.groupby('drv')\
            .agg(mean_hwy = ('hwy','mean'))
df_mpg
```

```python
df_mpg = mpg.groupby('drv', as_index = False)\
            .agg(mean_hwy = ('hwy','mean'))
df_mpg
```

```python
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

```python
df_mpg = df_mpg.sort_values('mean_hwy',ascending = False)
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

```python
mpg.head(10)
```

```python
df_mpg = mpg.groupby('drv', as_index = False)\
            .agg(n = ('drv','count'))
df_mpg
```

```python
sns.barplot(data = df_mpg, x = 'drv', y = 'n')
```

```python
cty_sort = mpg.query('category == "suv"')
cty_sort2 = cty_sort.groupby('model', as_index = False)\
        .agg(cty_mean=('cty','mean'))\
        .sort_values('cty_mean', ascending = False)\
        .head(5)
sns.barplot(data = cty_sort2, x = 'model', y = 'cty_mean')
```

```python
category = mpg.groupby('category', as_index = False)\
            .agg(count = ('category','count'))\
            .sort_values('count',ascending = False)
sns.barplot(data = category, x = 'category', y = 'count')
```
